**matplotlib** stateless approach

```python
np.random.seed(325)

rng = np.arange(50)
rnd = np.random.randint(0,10, size=(3, rng.size))
years = 1950 + rng

figure, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(years, rng+rnd, labels=["E", "W", "N"])
ax.set_title('Stateless')
ax.legend(loc='upper left')
ax.set_ylabel('Total')
ax.set_xlim(xmin=years[0], xmax=years[-1])
figure.tight_layout()

```

```python
x = np.random.randint(1, 11, size= 50)
y = x + np.random.randint(1,5, size=x.size)
data = np.column_stack((x,y))


figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_axisbelow(True)
ax1.grid(linestyle='--')

ax2.hist(data, bins=np.arange(data.min(), data.max()), label=('x','y'))
ax2.legend(loc=(0))
ax2.set_title('Frequencies of $x$ and $y$')
ax2.yaxis.tick_right()
ax2.set_axisbelow(True)
ax2.grid(linestyle='--')
```
